# Data Pipeline Validation
Validating feature distributions, normalization, correlations, and missing values.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# Mock data extraction representing the pipeline's output
np.random.seed(42)
df = pd.DataFrame(np.random.randn(1000, 5), columns=['RSI', 'MACD', 'Volume_Log', 'Ret_1d', 'Volatility'])

print('Missing Values:\n', df.isna().sum())
```

```python
# Feature distribution plots
df.hist(bins=50, figsize=(12, 8))
plt.suptitle('Feature Distributions')
```

```python
# Normalization verification (-1 to 1 check)
print('Min Values:\n', df.min())
print('Max Values:\n', df.max())

outliers = ((df < -1) | (df > 1)).sum()
print('Outliers outside [-1, 1]:\n', outliers)
```

```python
# Feature sanity checks and Correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Feature Correlation Matrix')
```
